```python
# qn 3:
# For double-precision floating point numbers:

under = 1.
over = 1.
for i in range(10000):
    under = under / 2.
    over = over * 2.
    print (over, under)
    
# from the code, the overflow happens at (8.98846567431158e+307) and underflow at (5e-324)
```

```python
# For single-precision floating point numbers:

under = np.float32(1.)
over = np.float32(1.)
for i in range(500):
    under = np.float32(under / 2.)
    over = np.float32(over * 2.)
    print (over, under)
    
# from the code, the overflow happens at (1.70141e+38) and underflow at (1.4013e-45)
```

```python
# For 32-bit, Intergers:

under = -np.int32(2**31-2)
over = np.int32(2**31-2)

for i in range(10):
    under = np.int32(under - 1)
    over = np.int32(over + 1)
    print (over, under)
    
# from the code, the overflow happens at (2147483647 --> -2147483648) and underflow at (-2147483648 --> 2147483647)
```

```python
import cmath

N = 10000
eps = (1.0)
for i in range (N):
    eps = (eps /2)
    Z = 1.0 + eps
    print ( "eps = " , eps ,  "Z =" , Z )
    
# from the code,the machine precision is (1.1102230246251565e-16)
```

```python
# for Complex numbers:

N = 10000
eps = complex (1.0j)
for i in range (N):
    eps = (eps /2)
    Z = complex (1.0j) + eps
    print ( "eps = " , eps ,  "Z =" , Z )
    
## from the code,the machine precision is (1.1102230246251565e-16j)
```
